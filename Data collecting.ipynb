```python
from csv import writer
import requests
from bs4 import BeautifulSoup


```

```python
result = requests.get("https://www.basketball-reference.com/leagues/NBA_2021_totals.html")
src = result.content
soup = BeautifulSoup(src, 'html.parser')

urls =  []
for tr_tag in soup.find_all("tr"):
    td_tag = tr_tag.find('td')
    if(td_tag!=None):
        a_tag = td_tag.find('a')
        urls.append(a_tag.attrs['href'])
```

# Create the categories for the Data frame

```python
#make the dictionary elements and category names
result = requests.get("https://www.basketball-reference.com/players/a/adamsst01.html")
src = result.content
soup= BeautifulSoup(src,'html.parser')

categories = []
table = soup.find('table')
thead = table.find("thead")
for th in thead.find_all("th")[1:]:
    categories.append(th.attrs["data-stat"])

categories.insert(0,"Name")
print(categories)
print("Categories: "+str(len(categories)))

#Create data frame
all_stats = {}
for i in categories:
    all_stats[i] = []


def addStats(list):
    for i in range(len(categories)):
        all_stats[categories[i]].append(list[i])

with open('NBA_Player_Data.csv', 'a') as f_object:
  
    # Pass this file object to csv.writer()
    # and get a writer object
    writer_object = writer(f_object)
  
    # Pass the list as an argument into
    # the writerow()
    writer_object.writerow(categories)
  
    #Close the file object
    f_object.close()
```
